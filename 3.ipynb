```python
import numpy as np
import pandas as pd

# Load and preprocess the dataset
file_path = 'data.csv'
data = pd.read_csv(file_path)

# Convert length and width to meters
def convert_to_meters(value, unit):
    if unit == 'centimeter':
        return value / 100
    return value

data['length_m'] = data.apply(lambda row: convert_to_meters(row['length'], row['lengthUnit']), axis=1)
data['width_m'] = data.apply(lambda row: convert_to_meters(row['width'], row['widthUnit']), axis=1)
data['area_sqm'] = data['length_m'] * data['width_m']
data['price_numeric'] = data['price'].str.replace('IDR ', '').str.replace(' Billion', '').astype(float)

# Select features and target variable
X = data[['area_sqm', 'isCarport']].values
y = data['price_numeric'].values

# Feature scaling
X_mean = X.mean(axis=0)
X_std = X.std(axis=0)
X_scaled = (X - X_mean) / X_std

# Add intercept term to features
X_scaled = np.c_[np.ones(X_scaled.shape[0]), X_scaled]

# Initialize parameters
beta = np.zeros(X_scaled.shape[1])

# Define learning rate and number of iterations
alpha = 0.01
iterations = 1000

# Gradient Descent
for _ in range(iterations):
    predictions = X_scaled @ beta
    errors = predictions - y
    gradient = X_scaled.T @ errors / len(y)
    beta -= alpha * gradient

# Coefficients
beta_0, beta_1, beta_2 = beta

# Predict function
def predict(area, is_carport):
    area_scaled = (area - X_mean[0]) / X_std[0]
    return beta_0 + beta_1 * area_scaled + beta_2 * is_carport

# Evaluate the model
predictions = X_scaled @ beta
mae = np.mean(np.abs(predictions - y))
mse = np.mean((predictions - y) ** 2)
r2 = 1 - (np.sum((predictions - y) ** 2) / np.sum((y - np.mean(y)) ** 2))

# Display results
print(f"Intercept (beta_0): {beta_0}")
print(f"Coefficient for area (beta_1): {beta_1}")
print(f"Coefficient for carport (beta_2): {beta_2}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

```python
import numpy as np
import pandas as pd

# Load the validation dataset
validation_file_path = 'validation_data.csv'
validation_data = pd.read_csv(validation_file_path)

# Preprocess the validation dataset similarly to the training dataset
validation_data['length_m'] = validation_data.apply(lambda row: convert_to_meters(row['length'], row['lengthUnit']), axis=1)
validation_data['width_m'] = validation_data.apply(lambda row: convert_to_meters(row['width'], row['widthUnit']), axis=1)
validation_data['area_sqm'] = validation_data['length_m'] * validation_data['width_m']
validation_data['price_numeric'] = validation_data['price'].str.replace('IDR ', '').str.replace(' Billion', '').astype(float)

# Select features and target variable from the validation dataset
X_val = validation_data[['area_sqm', 'isCarport']].values
y_val = validation_data['price_numeric'].values

# Feature scaling on validation data
X_val_scaled = (X_val - X_mean) / X_std
X_val_scaled = np.c_[np.ones(X_val_scaled.shape[0]), X_val_scaled]

# Predict using the trained model
y_val_pred = X_val_scaled @ beta

# Evaluate the model on the validation dataset
mae_val = np.mean(np.abs(y_val_pred - y_val))
mse_val = np.mean((y_val_pred - y_val) ** 2)
rmse_val = np.sqrt(mse_val)
r2_val = 1 - (np.sum((y_val_pred - y_val) ** 2) / np.sum((y_val - np.mean(y_val)) ** 2))

print(f"Mean Absolute Error (MAE): {mae_val}")
print(f"Mean Squared Error (MSE): {mse_val}")
print(f"Root Mean Squared Error (RMSE): {rmse_val}")
print(f"R-squared (R²): {r2_val}")
```
